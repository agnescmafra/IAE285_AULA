**CONVERSÃO**

* Definimos conversão de uma reação como o número de mols do reagente limitante reagidos em relação ao número de mols, deste, alimentados;

* Reações químicas normalmente são lentas, portanto não é fácil projetar um reator para a conversão completa do reagente limitante; 

* Então, a saída do reator contém reagentes não convertidos, que serão separados dos produtos e re-alimentam o reator (para os processos em fluxo contínuo).


\begin{align*} 
\\Conversão.de.um.reagente_X=\frac{{n}_{reageX}}{{n}_{alimentadoX}}=\frac{{n}_{entraX}-{n}_{saiX}}{{n}_{entraX}}\\
\\Conversão.global.de.um.reagente_X=\frac{{n}_{entra.no.processoX}-{n}_{sai.do.processoX}}{{n}_{entra.no.processoX}}\\
\\Conversão.por.passe.de.um.reagente_X=\frac{{n}_{entra.no.reatorX}-{n}_{sai.do.reatorX}}{{n}_{entra.no.reatorX}}\\
\end{align*}

**SELETIVIDADE**

* Também descreve o grau em que um reação de interesse predomina sobre as reações secundárias.

\begin{align*} 
\\Seletividade=\frac{{n}_{produto-desejado}}{{n}_{produto-indesejado-formado}}\\
\end{align*}

*Os cálculos do termo reage são realizados da seguinte maneira:*

* Como o limitante é o $C_3H_6$ o valor do alimentado é igual ao termo reage.
* Faz-se a regra de três : 


\begin{align*} 
\\1 mol de C_3H_6  \Longleftrightarrow  1 mols de NH_3\\
\\10 mols de C_3H_6 \Longleftrightarrow  x mols de NH_3\\
\end{align*} 

```python
import sympy
from sympy.solvers import solve
from sympy import Symbol
import math
x = Symbol ('x')
sol1 = solve (1*x-10*1,x)
print('Nreage de NH3 =', sol1)
```

\begin{align*} 
\\1 mol de C_3H_6  \Longleftrightarrow  3/2 mols de O_2\\
\\10 mols de C_3H_6 \Longleftrightarrow  y mols de O_2\\
\end{align*} 

```python
import sympy
from sympy.solvers import solve
from sympy import Symbol
y = Symbol ('y')
sol2 = solve (1*y-10*(3/2),y)
print('Nreage de O2 =', sol2)
```

Cálculo dos Excessos:

\begin{align*} 
\\Excesso_A=\frac{{n}_{alimentadoA}-{n}_{reageA}}{{m}_{reageA}}\\
\end{align*}

```python
Excesso_NH4 = 100*(12-10)/10
print(Excesso_NH4)
```

```python
Excesso_O2 = 100*(78-15)/15
print(Excesso_O2)
```

Quanto será a composição dos produtos?

\begin{align*} 
\\ C_3H_6 + NH_3 + 3/2 O_2 \Rightarrow C_3H_3N + 3 H_2O
\end{align*}

**Equação Geral**

***ENTRA - SAI + GERADO - CONSUMO = ACÚMULO***

Para os REAGENTES : SAI = ENTRA - CONSUMO

Para os PRODUTOS : SAI = ENTRA + REAGE


* Vamos fazer os cálculos para os reagentes primeiro ???


```python
#Para os REAGENTES : SAI = ENTRA - CONSUMO

#N_final_C3H6 = N_alimentado_C3H6 - N_reage_C3H6
N_final_C3H6 = 10 - 10
print('N_final_C3H6 =', N_final_C3H6)

#N_final_NH3 = N_alimentado_NH3 - N_reage_NH3
N_final_NH3 = 12 - 10
print('N_final_NH3 =',N_final_NH3)

#N_final_O2 = N_alimentado_O2 - N_reage_O2
N_final_O2 = 78 - 37
print('N_final_O2 =',N_final_O2)
```

**Fundamentos do Balanço de Massa**

*   Lei de conservação de matéria​;
*   Sempre há pequenas perdas de massa ao longo do processo​;
*   Faz-se em bases atômicas, moleculares ou gerais;
*   Deve-se escolher onde fazer o balanço: SISTEMA e suas vizinhanças;
*   Volume de controle: volume definido de interesse para análise de balanço;
*   Superfície de controle: fronteira ou superfície que envolve o controle e por ela passa massa ao volume.


**Equação Geral**

***ENTRA - SAI + GERADO - CONSUMO = ACÚMULO***

**Classificação dos Processos**

Tempo
*   Estado-estacionário;
*   Transiente;

Espaço
*   Contínuo;
*   Batelada;
*   Semi-contínuo;


**Estado estacionário ou regime permanente:​**

*   As principais variáveis de processo não variam com o tempo, exceto pequenas flutuações.​
 
**Transiente:**​

*   Há variações com o tempo das principais variáveis de processo.

*   Um processo pode apresentar variáveis em transição com o tempo e outras em regime permanente.

*   A tendência dos processos é serem operados em regime permanente, mas na partida ou em mudanças de set-point há transiência.

**Processos em Batelada**
---


*   Alimentação é carregada no sistema num dado estado inicial e descarregado num outro dado momento;
*   Não há transferência de massa através da superfície de controle nesse período.

**Equacionamento para batelada sem reação:**

$$\mbox{Acúmulo} = \mbox{Entrada} - \mbox{Saída} + \underbrace{\mbox{Gerado}}_{=0} - \underbrace{\mbox{Consumo}}_{=0}$$
\begin{align*} 
\\Entrada = \dot{m}_{entra} = 0 \\
\\Saída = \dot{m}_{saída} = 0 \\
Acúmulo = \frac{dm}{dt}=0\\
\\Integrando :\frac{dm}{dt}=0
\\m_{entra}=m_{saída}
\end{align*}

**Processos Contínuos**
---


*   Ocorre passagem contínua de matéria pela superfície de controle.


**Equacionamento processo contínuo sem reação:**

$$\mbox{Acúmulo} = \mbox{Entrada} - \mbox{Saída} + \underbrace{\mbox{Gerado}}_{=0} - \underbrace{\mbox{Consumo}}_{=0}$$
\begin{align*} 
\\Entrada = \dot{m}_{entra}\\
\\Saída = \dot{m}_{saída}\\
\frac{dm}{dt}=0\\
\dot{m}_{entra}= \dot{m}_{saída}\\
\end{align*}

**Processos Semi-Contínuos ou Semi-batelada**
---


*   A entrada/saída de material é instantânea e a saída/entrada é contínua. Por exemplo: o processo de enchimento de uma caixa d'água. 


**Equacionamento processos semi-contínuos sem reação:**

$$\mbox{Acúmulo} = \mbox{Entrada} - \mbox{Saída} + \underbrace{\mbox{Gerado}}_{=0} - \underbrace{\mbox{Consumo}}_{=0}$$
\\begin{align*} 
\\Entrada = \dot{m}_{entra}\\
\\Saída = \dot{m}_{saída}\\
\frac{dm}{dt}\neq 0\\
\dot{m}_{entra}= 0\\ou\\\dot{m}_{saída} = 0\\
\end{align*}

**Definindo fração mássica e fração molar**
---

*Você irá perceber que em muitos processo industriais em uma mesma corrente, entrará mais de uma substância. Mas o nosso papel de engenheiro é entender e controlar todas as substâncias que entram e saem do volume de controle.* 

*Para tal, a definição de fração mássica e fração molar, junto com a Lei da Conservação da Massa nos ajuda a montar os balanços de massa e entender o que está acontecendo no processo industrial.*

*Fazer inventário é o nosso negócio, não se esqueça!*

*Hipotéticamente se estivermos trabalhando em um processo industrial com duas substâncias A e B, sem reação, que entram por uma mesma corrente de entrada e saem por uma mesma corrente de saída, podemos propor as seguintes equações para a definição de fração mássica e fração molar.*

**Fração mássica da substância A e B na corrente de entrada:**
\begin{align*} 
\\x_{A.ENTRADA} =\frac{{m}_{A.correnteEntrada}}{{m}_{Total.correnteEntrada}}\\
\\x_{B.ENTRADA} =\frac{{m}_{B.correnteEntrada}}{{m}_{Total.correnteEntrada}}\\
\end{align*}
**Fração mássica da substância A e B na corrente de saída:**
\begin{align*} 
\\x_{A.SAÍDA} =\frac{{m}_{A.correnteSaída}}{{m}_{Total.correnteSaída}}\\
\\x_{B.SAÍDA}  =\frac{{m}_{B.correnteSaída}}{{m}_{Total.correnteSaída}}\\
\end{align*}

\begin{align*} 
\\x_{A.ENTRADA} + x_{B.ENTRADA}=1 
\\x_{A.SAÍDA} + x_{B.SAÍDA}=1 
\end{align*}

**Fração molar da substância A e B na corrente de entrada:**
\begin{align*} 
\\y_{A.ENTRADA} =\frac{{n}_{A.correnteEntrada}}{{n}_{Total.correnteEntrada}}\\
\\y_{B.ENTRADA} =\frac{{n}_{B.correnteEntrada}}{{n}_{Total.correnteEntrada}}\\
\end{align*}
**Fração molar da substância A e B na corrente de saída:**
\begin{align*} 
\\y_{A.SAÍDA} =\frac{{n}_{A.correnteSaída}}{{n}_{Total.correnteSaída}}\\
\\y_{B.SAÍDA}  =\frac{{n}_{B.correnteSaída}}{{n}_{Total.correnteSaída}}\\
\end{align*}

\begin{align*} 
\\y_{A.ENTRADA} + y_{B.ENTRADA}=1 
\\y_{A.SAÍDA} + y_{B.SAÍDA}=1 
\end{align*}

**By-pass, Reciclo e Purga**
---

**BY-PASS** : Operação de desvio parte da alimentação de uma unidade de processo, para se unir mais a frente a outra corrente, seja para diluir componentes na corrente, ou atingir alguma especificação no projeto/processo. Surgem 4 volumes de controle, nem todos são necessários balanços de massa para serem resolvidos​.

**RECICLO** : Nem sempre é possível reagir completar a reação de um componente, ou separá-lo e isso significa prejuízo se não for reaproveitado. Para resolver isso, desvia-se parte de uma corrente para ser reprocessado.​
* Catalisadores também são reciclados para não causar problemas adiante no processo;
* Diluir uma corrente de processo-lamas de cristalizadores/evaporadores​;
* Controle de uma variável de processo – temperatura de reação​;
* Circurlação de um fluido de trabalho – gás de refrigeração.

**PURGA** : Prática de descartar parte de uma corrente para evitar o acúmulo de compostos inertes ou não desejados. Em processo com reciclo é utilizado para que o balanço de massa feche e não acumule matéria.
